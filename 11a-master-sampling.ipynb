# 3. Resample each sale.csv and combine them into a master dataframe

### *Note*
### The total number of tire sales from 2015-2018 is 29254956, and we want to have 100000 samples in our master sample data. The ratio calculation is 100000/29254956 = 0.00341822424. 

```python
master_df = pd.DataFrame()
file_names = glob.glob('/data/p_dsi/teams2023/team7/*tire_sales*.csv')

for file_name in file_names:
    file = pd.read_csv(file_name, sep = ",")
    file_sampled = file.sample(frac=0.00341822424, replace=False, random_state=1)
    master_df = pd.concat([master_df, file_sampled])
```

```python
master_df.to_csv("/data/p_dsi/teams2023/team7/sales_sample.csv")
```

# 4. Join sales_sample.csv with vehicle.csv, store.csv, individual.csv and write master_sample.csv

## 4.1 Read in data

```python
tire_sales_sampled = pd.read_csv("/data/p_dsi/teams2023/team7/sales_sample.csv")
```

```python
individual = pd.read_csv("/data/p_dsi/teams2023/bridgestone_data/data/individual.csv")
# product read before
store = pd.read_csv("/data/p_dsi/teams2023/bridgestone_data/data/store.csv", sep = "|")
vehicle = pd.read_csv("/data/p_dsi/teams2023/bridgestone_data/data/vehicle.csv", sep = "|")
```

## 4.2 Merge

```python
final_df = tire_sales_sampled.merge(individual, left_on = 'INDIV_ID', right_on = "MZB_INDIV_ID", how = 'left')
```

```python
final_df = final_df.merge(product, on = "ARTICLE_ID", how = 'left')
```

```python
final_df = final_df.merge(store, on = "STORE_ID", how = 'left')
```

```python
final_df = final_df.merge(vehicle, on = "VEHICLE_ID", how = 'left')
```
