```python
import pandas as pd
import glob
```

# Data Validation for sales_2015

```python
csv_files = glob.glob('/data/p_dsi/teams2023/bridgestone_data/data/*sales_2015*.csv')
```

```python
df = pd.read_csv("/data/p_dsi/teams2023/bridgestone_data/data/sales_20150430.csv", sep = "|")
```

```python
ref_columns = set(df.columns)
print(ref_columns)
```

```python
for df in csv_files:
    df = pd.read_csv(df, sep = "|")
    if not set(df.columns) == ref_columns:
        print(f"Data frame {df} does not have the same columns as the reference set.")
    else:
        print("Columns are identical!")
```
