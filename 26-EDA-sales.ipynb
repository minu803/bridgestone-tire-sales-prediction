```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
master_sales = pd.read_csv("/data/p_dsi/teams2023/team7/final_sales.csv")
```

# Units Distribution

```python
master_sales["UNITS"].unique()
```

```python
units_summary = pd.DataFrame(master_sales.groupby("UNITS").size(), columns = ["count"])
units_summary.plot(kind = "bar")

# Set the size of the display window
fig = plt.gcf()
fig.set_size_inches(15, 5)
plt.ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```

```python
units_summary_sorted = pd.DataFrame(master_sales.groupby("UNITS").size(), columns = ["count"]).sort_values(by=['count'], ascending = False)
units_summary_sorted.plot(kind = "bar")

# Set the size of the display window
fig = plt.gcf()
fig.set_size_inches(15, 5)
plt.ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```

```python
years = ["2015", "2016", "2017", "2018"]
for year in years:
    path = "/data/p_dsi/teams2023/team7/final_sales_" + year + ".csv"
    sales = pd.read_csv(path)
    
    sales["DATE"] = pd.to_datetime(sales['DATE'])
    sales['MONTH'] = sales['DATE'].dt.month
    
    # Group by month and category and sum units
    grouped_df = sales.groupby(['MONTH', 'ARTICLE_ID'])['UNITS'].sum().reset_index()

    # Pivot the dataframe to get category as columns
    pivot_df = grouped_df.pivot(index='MONTH', columns='ARTICLE_ID', values='UNITS')

    # Plot the resulting DataFrame as a bar plot
    pivot_df.plot(kind='bar', position=0.5, width=0.9)

    # Add labels and title to the plot
    plt.xlabel('Month')
    plt.ylabel('Total Units')
    plt.title('ARTICLE_ID in ' + year)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
#     plt.ylim(0, 200000)

    # Show the plot
    plt.show()
```
